```python
import numpy as np
from matplotlib import pyplot as plt
import math
from numpy import linalg as lg
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
plt.plot([f(x) for x in np.arange(1,15,0.1)])
```

```python
def poly_matrix(dots, f):
    A = np.ones((len(dots), len(dots) + 1))
    for i in range(len(dots)):
        for j in range(len(dots)):
            A[i][j] = dots[i]**j
        A[i][len(dots)] = f(dots[i])
    return A
```

```python
def poly(coef):
    return lambda x: sum([x**i*coef[i] for i in range(len(coef))])
```

```python
dots = [1, 15]
A = poly_matrix(dots, f)
coef = lg.solve(A[:, :len(dots)], A[:, len(dots):])
plt.plot([poly(coef)(x) for x in np.arange(0, 16, 0.1)])
```

```python
dots = [1, 8, 15]
A = poly_matrix(dots, f)
coef = lg.solve(A[:, :len(dots)], A[:, len(dots):])
plt.plot([poly(coef)(x) for x in np.arange(0, 16, 0.1)])
```

```python
dots = [1, 4, 10, 15]
A = poly_matrix(dots, f)
coef = lg.solve(A[:, :len(dots)], A[:, len(dots):])
plt.plot([poly(coef)(x) for x in np.arange(0, 16, 0.1)])
```
