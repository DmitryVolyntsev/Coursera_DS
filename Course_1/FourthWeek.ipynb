```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Экспоненциальное распределение

```python
exp_rasp = sts.expon()
sample = exp_rasp.rvs(1000)       #выборка объема 1000
```

### Строим график теоретической плотности распределения и гистограмму

```python
# плотность распределения
x = np.linspace(0,6,100)
pdf = exp_rasp.pdf(x)
plt.plot(x, pdf, label='Теоретическая pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

# гистограмма
plt.hist(sample, bins = 20, density=True)        #density == normed in Python 3
plt.ylabel('Доля')
plt.xlabel('$x$')
```

### Матожидание и дисперсия для экспоненциального распределения

$\mathbb{E} = \frac{1}{\lambda} = 1$

$\mathbb{D} = \frac{1}{\lambda^2} = 1$

### Построим гистограммы и плотности распределений выборочных средних

```python
E = 1
D = 1
sigma = 1
```

1) n = 5

```python
x = np.linspace(0,3,100)
sigma_5 = math.sqrt(D/5)                       #отклонение, соответствующее новой дисперсии
midchoice_list_5 = [0 for i in range(1000)]    #1000 выборок объема 5
for i in range(1000):
    sample_5 = exp_rasp.rvs(5)
    midchoice_list_5[i] = sum(sample_5)/5      #массив из 1000 выборочных средних
norm_rasp_5 = sts.norm(E, sigma_5)             #соответствующее нормальное распределение
pdf_5 = norm_rasp_5.pdf(x)
plt.plot(x, pdf_5, label='Теоретическая pdf', alpha=0.5)
plt.hist(midchoice_list_5, bins = 30, density=True)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
